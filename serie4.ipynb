## 1) Explique como você pode obter o conjunto dos graus associado a cada um dos dos nós do grafo.

## Implementação em matriz de adjacência:

```python
G = AdjGraph({'a', 'b', 'c', 'd', 'e'})
G.add_edge('a', 'b')
G.add_edge('a', 'e')
G.add_edge_from([('e', 'c'), ('b', 'c'), ('a', 'c')])
```

Para determinar o grau do vértice 'a' basta checar a quantidade de 1's em sua respectiva linha na matriz de adjacência:

```python
G.print_sorted()
```

Na linha a temos 3 1's => grau 3 

## <b>Implementação em lista de adjacência:</b>


```python
G2 = ListGraph({'a', 'b', 'c', 'd', 'e'})
G2.add_edge('a', 'b')
G2.add_edge('a', 'e')
G2.add_edge_from([('e', 'c'), ('b', 'c'), ('a', 'c')])
```

Para determinar o grau do vértice 'b' basta contar a quantidade de nós armazenados no nó 'b' da lista gerada pelo grafo:

```python
G.print_sorted()
```

Dois nós estão armazenados no nó 'b' => grau = 2

```python
G3 = incGraph({'a', 'b', 'c', 'd', 'e'})
G3.add_edge('a', 'b', 'A')
G3.add_edge('a', 'e', 'B')
G3.add_edge_from([('e', 'c', 'C'), ('b', 'c', 'D'), ('a', 'c', 'E')])
```

Para determinar o grau do nó 'e' basta verificar a quantidade de 1's na sua linha correspondente da matriz 

```python
G.print_sorted()
```

Temos 2 1's na linha 'e' da matriz, logo sabemos que o grau é 2

## Usando uma implementação computacional de um algoritmo, para cada número inteiro do 1,500, ache quantos e quais seus divisores. Faça um gráfico com os resultados que você achou.

```python
def divisors(num):
    alldiv = []
    for x in range (1, int(num/2)+1):
        if(num % x) == 0:
            alldiv.append(x)
    alldiv.append(num)
    return alldiv
```

```python
import math
```

```python
def all_divisors(a, b):
    divlist = []
    for n in range(a, b+1):
        divlist.append(divisors(n))
    return divlist
```

```python
divisores = all_divisors(1,1500)
```

```python
import matplotlib.pyplot as plt
x = list(range(1, len(divisores)+1))
for xi, yi in zip(x, divisores):
    plt.scatter([xi] * len(yi), yi, s=3)
```
